```python
import os
import pandas as pd
from tqdm import tqdm

DATA_DIR = "/kaggle/working/data/"
os.makedirs(DATA_DIR, exist_ok=True)
```

```python
def optimize_df(df):
    for col in df.columns:
        if df[col].dtype == "float64":
            df[col] = df[col].astype("float32")
        elif df[col].dtype == "int64":
            df[col] = df[col].astype("int32")
    return df
```

```python
def process_to_parquet_files(folders, out_dir, prefix):
    os.makedirs(out_dir, exist_ok=True)

    counter = 0

    for folder in folders:
        if not os.path.exists(folder):
            print("Missing:", folder)
            continue

        files = sorted([os.path.join(folder, f) for f in os.listdir(folder) if f.endswith(".csv")])

        print(f"\n📂 {folder} → {len(files)} files")

        for file in tqdm(files):
            try:
                df = pd.read_csv(file)
                if df.empty:
                    continue

                df = optimize_df(df)

                out_path = os.path.join(out_dir, f"{prefix}_{counter:06d}.parquet")

                df.to_parquet(out_path, engine="pyarrow", compression="snappy", index=False)

                counter += 1

            except Exception as e:
                print(f"❌ {file}: {e}")
```

```python
train_files = [os.path.join(DATA_DIR + "train_2017", f) for f in os.listdir(DATA_DIR + "train_2017")]
test_files  = [os.path.join(DATA_DIR + "test_2016", f) for f in os.listdir(DATA_DIR + "test_2016")]

df_train = pd.concat([pd.read_parquet(f) for f in train_files], ignore_index=True)
df_test  = pd.concat([pd.read_parquet(f) for f in test_files], ignore_index=True)

print(df_train.shape, df_test.shape)
```

```python
def analyze_columns(df):
    results = []

    for col in df.columns:
        n_unique = df[col].nunique()

        if pd.api.types.is_numeric_dtype(df[col]):
            min_val = df[col].min()
            max_val = df[col].max()
        else:
            min_val = None
            max_val = None

        results.append({
            "column": col,
            "n_unique": n_unique,
            "min": min_val,
            "max": max_val
        })

    return pd.DataFrame(results)
```

```python
analyze_columns(df_train)
```

```python
def detect_constant_columns(df):
    constant_cols = []
    informative_cols = []

    for col in df.columns:
        n_unique = df[col].nunique()

        if n_unique == 1:
            constant_cols.append(col)
        else:
            if pd.api.types.is_numeric_dtype(df[col]):
                if df[col].min() == df[col].max():
                    constant_cols.append(col)
                else:
                    informative_cols.append(col)
            else:
                informative_cols.append(col)

    return constant_cols, informative_cols
```

```python
 detect_constant_columns(df_train)
```

```python
def select_smart_columns(df, keyword="smart"):
    return [col for col in df.columns if keyword in col.lower()]
```

```python
select_smart_columns(df_train, keyword="smart")
```

```python
smart_cols = select_smart_columns(df_train)

df_smart = df_train[smart_cols]

constant_cols, informative_cols = detect_constant_columns(df_smart)
```

```python
print("Colonnes constantes :", len(constant_cols))
print("Colonnes informatives :", len(informative_cols))
```
